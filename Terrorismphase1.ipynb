```python
f,ax = plt.subplots(figsize=(15,15))
ax = sns.barplot(x="Count", y="Nation", data=nationtype[0:50],
                 palette="cubehelix").set_title('Nations That Suffered Most of the TERRORISM Attacks')
```

```python
plt.figure(figsize=(5,5))
target_counts = dataset['targtype'].value_counts()
colors = [
    "#000004", "#070b39", "#1c1e6d", "#414387", "#5a60a0",
    "#7577b9", "#8f8fcc", "#a9a8e6", "#c4c6f1", "#dfdffa",
    "#f1ebf6", "#f9f4f8", "#fbebf5", "#f9dbdb", "#f4bccf",
    "#ec90c7", "#e763bc", "#dd3497", "#d7306e", "#ae0e36"]
```

```python
plt.figure(figsize=(4, 4))

provstate_counts = dataset['provstate'].value_counts().nsmallest(10)
pie_colors = ["#ec90c7", "#e763bc", "#dd3497", "#d7306e", "#ae0e36","#f1ebf6", "#7577b9", "#8f8fcc", "#a9a8e6", "#c4c6f1", "#dfdffa", "#fbebf5", "#f9dbdb", "#f4bccf" ]

wedges, texts, autotexts = plt.pie(provstate_counts, labels=provstate_counts.index, colors=pie_colors, autopct='%1.1f%%')


for text in texts + autotexts:
    text.set_fontsize(6)

plt.title("Distribution of Attacks in Top 10 States")

plt.show()
```

```python
f, ax = plt.subplots(figsize=(15, 10))
ax = sns.barplot(x="Counts", y="Groupname", data=nametype[:10],palette="RdBu").set_title('Terrorist Groups')
```

```python
plt.figure(figsize=(7, 5))
top_cities = dataset['city'].value_counts().nsmallest(10).index
subset = dataset[dataset['city'].isin(top_cities)]
heatmap_data = subset.pivot_table(index='city', columns='year', values='attacktype1', aggfunc='sum')

sns.heatmap(heatmap_data, cmap="icefire")

plt.title("Attack Counts in Top 10 Cities with Fewest Attacks")
plt.xlabel("Year")
plt.ylabel("City")
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(15, 5))
sns.countplot(data=dataset, y='attacktype', order=dataset['attacktype'].value_counts().index, palette='viridis')
plt.title('Distribution of Attack Types')
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(10,8))
sns.scatterplot(x = "latitude",y = "longitude",data=dataset[0:2000],alpha=1,color='black')
plt.title('Terrorist Attack Locations',fontsize=15)
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.show()
```

```python
plt.figure(figsize=(10, 6), dpi=100)

sns.barplot(
    x=dataset['provstate'].value_counts()[:15].index,
    y=dataset['provstate'].value_counts()[:15].values,
    palette="icefire")

plt.title("Top 15 areas that are attacked")
plt.xlabel("Area")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
